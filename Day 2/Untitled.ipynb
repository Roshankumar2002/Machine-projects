```python
x=np.array([[1.8],[1.8]])
m1 = np.array([[1],[1]])
m2 = np.array([[3],[3]])
s=np.array([[1,0],[0,1]])
Pw1 = 0.5
Pw2 = 0.5
```

```python
e1 = -0.5 * np.dot(np.dot(np.subtract(x,m1).transpose(), np.linalg.inv(s)), np.subtract(x,m1))
```

```python
e1=e1[0][0]
```

```python
px1 = (1/(2*np.pi * np.sqrt(np.linalg.det(s))))*np.exp(e1)
```

```python
e2 = -0.5 * np.dot(np.dot(np.subtract(x,m2).transpose(),np.linalg.inv(s)), np.subtract(x,m2))
```

```python
e2 = e2[0][0]
```

```python
px2 = (1/(2*np.pi*np.sqrt(np.linalg.det(s))))*np.exp(e2)
```

```python
p1 = px1*Pw1
p2 = px2*Pw2
print(f"p1 = {p1}, p2={p2}")
if(p1>p2):
    print("X is in class 1")
else:
    print("X is in class 2")
```

# Q. no 1

```python
import numpy as np

# Define the means and covariance matrix
m1 = np.array([0, 0, 0])
m2 = np.array([0.5, 0.5, 0.5])
S = np.array([[0.8, 0.01, 0.01], [0.01, 0.2, 0.01], [0.01, 0.01, 0.2]])
x = np.array([0.1, 0.5, 0.1])

# Euclidean Distance Classifier
def euclidean_distance(x, m):
    return np.sqrt(np.sum((x - m)**2))

# Mahalanobis Distance Classifier
def mahalanobis_distance(x, m, S):
    S_inv = np.linalg.inv(S)
    return np.sqrt(np.dot(np.dot((x - m).transpose(), S_inv), (x - m)))

# Classify the point x
euclidean_dist_m1 = euclidean_distance(x, m1)
euclidean_dist_m2 = euclidean_distance(x, m2)
mahalanobis_dist_m1 = mahalanobis_distance(x, m1, S)
mahalanobis_dist_m2 = mahalanobis_distance(x, m2, S)

print(f"Euclidean Distance to m1: {euclidean_dist_m1}")
print(f"Euclidean Distance to m2: {euclidean_dist_m2}")
print(f"Mahalanobis Distance to m1: {mahalanobis_dist_m1}")
print(f"Mahalanobis Distance to m2: {mahalanobis_dist_m2}")

# Determine the class based on the smaller distance
class_euclidean = 1 if euclidean_dist_m1 < euclidean_dist_m2 else 2
class_mahalanobis = 1 if mahalanobis_dist_m1 < mahalanobis_dist_m2 else 2

print(f"Classified to Class {class_euclidean} using Euclidean distance.")
print(f"Classified to Class {class_mahalanobis} using Mahalanobis distance.")
```

# Q. no 2

```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the means and covariance matrices
m1 = np.array([1, 1])
m2 = np.array([3, 3])
S1 = np.array([[0.1, -0.08], [-0.08, 0.1]])
S2 = np.array([[0.2, 0.05], [0.05, 0.2]])
```

## Part 1

```python

# Prior probabilities
P1 = 0.5
P2 = 0.5

# Generate points for each distribution
def generate_points(m, S, P, N):
    return np.random.multivariate_normal(m, S, size=int(P*N))

# Generate the set X
N = 500
points_class_1 = generate_points(m1, S1, P1, N)
points_class_2 = generate_points(m2, S2, P2, N)
X = np.vstack((points_class_1, points_class_2))

# Plot the points
plt.scatter(X[:, 0], X[:, 1], alpha=0.5)
plt.title('2-Dimensional PDF')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [[0,0],[0,1],[1,0],[1,1]]
x_array = np.asarray(x)
out = x_array[:,1]* x_array[:,0]
out
```

```python
w = np.random.rand(1,3) * 10
w_1 = 0.5
w_2 = 0.5
b = 0
```

```python
def step(net):
    if net>=0:
        return 1
    else: 
        return 0
    
```

```python
error = np.array([0,0,0,0])
for i in range(len(x)):
    f_net = step(np.dot(np.asarray([w_1,w_2]), x[i]) +b)
    print(f_net)
    error[i] = out[i] - f_net
    print(error)
E= np.sum(error)
E
```

```python
plt.figure(figsize = (4,4))
plt.scatter(x_array[:,0],x_array[:,1],c=out)
m=np.linspace(0,1.4)
n=(-w_1 * m -b)/w_2
plt.plot(m,n,'-r')
# plt.ylim([-0.5,1.5])
# plt.xlim([-0.5,1.5])
plt.grid(True)
plt.show()
```

```python
max_it = 1000
t=1
learning_rate = 0.1
vals=[[w_1,w_2,b]]
while t<max_it & E!=0:
    for i in range(len(x)):
        f_net = step(np.dot(np.asarray([w_1,w_2]),x[i])+b)
        error[i] = out[i] - f_net
        print(error[i])
        w_1 = w_1 + learning_rate * error[i] * x[i][0]
        w_2 = w_2 + learning_rate * error[i] * x[i][1]
        b = b + learning_rate * error[i]
    vals.append([w_1,w_2,b])
    E = np.sum(error)
    t=t+1
```

```python
plt.figure(figsize =(4,4))
plt.scatter(x_array[:,0],x_array[:,1],c=out)
m = np.linspace(0,1.4)
n = (-w_1*m - b) / w_2
plt.plot(m,n,'-r')
plt.grid(True)
plt.show()
```
